```python
import pandas as pd 
import numpy as numpy
```

```python
#Load both CSVs
products=pd.read_csv("/Users/camiloandres/Documents/data_analysis_projects/ecommerce_sales_analysis/data/products.csv")
orders=pd.read_csv("/Users/camiloandres/Documents/data_analysis_projects/ecommerce_sales_analysis/data/orders.csv")
```

## Inspection of the Datasets

```python
#Shape and info 
print(products.shape)
print(products.info())
print(orders.shape)
print(orders.info())

#Missing values 
print(products.isnull().sum())
print(orders.isnull().sum())

#Duplicates
print(products.duplicated().sum())
print(orders.duplicated().sum())



```

## Validate Relationships

```python
#Check if all order product_ids exist in products
invalid_products=orders[~orders["product_id"].isin(products["product_id"])]
print(invalid_products)
```

The relationshionship between orders.product_id and products.product_id es clean and valid, which means that every order corresponds to a real product. 

## Convert and format data types 

```python
orders["order_date"]=pd.to_datetime(orders["order_date"],errors="coerce")
orders["product_id"]=orders["product_id"].astype(str)
products["product_id"]=products["product_id"].astype(str)
```

## Merge the tables 

```python
merged_df = pd.merge(
    orders,
    products,
    on='product_id',
    how='left'
)
```

```python
#Check
merged_df.info()
merged_df.head()
```

## Export cleaned data

```python
merged_df.to_csv('../data/cleaned_merged_data.csv', index=False)
print("Cleaned data exported successfully!")
```
